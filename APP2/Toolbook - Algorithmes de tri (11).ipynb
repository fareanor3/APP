## Algorithme de tri

### Définition

Algorithme qui, partant d’une liste, renvoie une version **ordonnée** de la liste.

Trier des données consiste à les ranger en ordre croissant ou décroissant.

On peut trier différents types de données. Le tout est de définir un critère de comparaison pour ces données.

Exemple :

- Personnes par ordre alphabétique,
- Joueurs par score,
- Fichiers par taille,
- Fichiers par date d’édition, 
- Dossier par profondeur dans une arborescence, 
- ... etc.

Il existe de nombreux algorithmes de tri. Parmi lesquels on peut citer :

- Tri par insertion
- Tri par selection
- Tri à bulle
- Tri fusion
- Tri rapide
- Tri par tas

Pourquoi tous ces algorithmes ? parceque les applications sont variées.
- Si on a une application avec beaucoup de données, on cherchera à limiter **la complexité calculatoire**.
- Si on a une application avec des données qui occupent beaucoup d’espace, on cherchera cette fois ci à limiter **la complexité spatiale**.

### Tri et complexité

Une opération de tri consomme un temps de calcul important sur un ordinateur. Il est donc nécessaire d’étudier la complexité temporelle des différents algorithmes de tri. On peut évaluer cette complexité  dans  le  "pire  des  cas",  ou  dans  le  "meilleur  des  cas"  ou "en  moyenne". On pourra ainsi choisir l’algorithme qui convient le mieux suivant la situation. Mais généralement quand on parle de complexité temporelle, on parle de complexité dans le pire des cas.

La complexité temporelle mesure le nombre d'opérations élémentaires effectuées pour trier un ensemble de données. C'est un critère très important pour comparer les algorithmes de tri, puisqu'il représente une estimation directe du temps d'exécution de l'algorithme.

Les algorithmes étudiés sont basés sur des comparaisons successives entre les éléments. De ce fait la complexité d’un algorithme a le même ordre de grandeur que le nombre de comparaisons effectuées par cet algorithme.  Il est à noter que la comparaison et l'échange éventuel de deux éléments s'effectuent en temps constant.

La complexité en temps est souvent notée T et est exprimée comme une fonction du nombre n d'éléments à trier à l'aide de la notation de Landau $O$. La notation grand $O$ de Landau (en anlgaos Big $O$) décrit une borne supérieure. On peut donc l'utiliser pour obtenir une borne supérieure sur le temps d'exécution dans le pire des cas pour des données quelconques.

On dispose ainsi de plusieurs algorithmes qui se distinguent par leurs complexités temporelles. Dans ce toolbook, nous allons illustrer quelques-uns de ces algorithmes avec des listes ou des tableaux.




## Tri par insertion

En anglais :insertion sort.


Ce tri est connu comme le tri du joueur de carte. Initialement, vous tenez toutes vos cartes (non triées) dans votre main. Suposons que vous triez de gauche à droite. À chaque étape du tri, vous prenez la “première la plus à gauche dans "le tas à trier" qui se trouve dans votre main, et l’insérez à sa place dans "le tas gauche déjà trié". À la fin, vos cartes sont triées par ordre croissant dans votre main.

Dans l'algorithme, on parcourt le tableau à trier du début à la fin. L'objectif d'une étape est d'insérer le i-ème élément à sa place parmi ceux qui le précèdent. Il faut pour cela trouver où l'élément doit être inséré en le comparant aux autres qui le précèdent, puis décaler les éléments en remontant dans la liste jusqu’à trouver la bonne place et effectuer l'insertion à sa bonne place. 


Considérons maintenant une liste ou un tableau  dont les premiers éléments d’indices 0, 1, ..., i -1 sont déjà triés. 
On cherche à placer l’élément d’indice i à sa bonne place parmi les éléments déjà triés, pour cela on le compare aux précédents et on le décale jusqu’à ce que sa place T[i]soit trouvée.

Ainsi on itère sur la liste:
- Chaque nouvel élément est inséré à sa place parmi les éléments déjà triés .
- On recommence jusqu’à épuisement.

Exemple :

Voici une animation représentant le déroulement du tri par insertion sur la liste [6, 5, 3, 1, 8, 7, 2, 4] ainsi que les étapes de l'exécution de l'algorithme à chaque itération i.







```python
def insertion_sort(array):
    n = len(array)
    for i in range(1, n):
        key = array[i]
        j = i - 1
        while j >= 0 and array[j] > key:
            array[j + 1] = array[j]
            j = j - 1
        array[j + 1] = key
        print(array)
    return array

array_insertion=insertion_sort([76,32,65,2,80,11,7])        
```

### Complexité du tri par insertion

Calculons la complexité dans le pire des cas de ce tri :
- Dénombrons les comparaisons entre éléments de la liste. Dans tous les cas on a n-1 passages dans la boucle for. Dans la boucle while, le nombre d’itérations est maximal lorsque la liste est triée dans l’ordre décroissant, ce qui conduit à i-1 comparaisons. Soit en tout, le temps d’exécution est donné par:

$$ T(n)= \sum_{i=1}^{n-1}(i) = 1+2+⋯+𝑛−1 = \frac{n(n-1)}{2} \simeq \frac{n^{2}}{2} $$

La complexité en temps de ce tri est en $O(n^{2})$ dans le pire des cas (cas où la liste est triée dans l’ordre inverse et donc chaque clé doit être comparée à tous les éléments précédents) et en $O(n)$ dans le meilleur des cas (cas où la liste est déjà triée avant application de l’algorithme). On peut également vérifier que le tri est en moyenne en $O(n^{2})$.

## Tri par sélection

En anglais : selection sort.

Ce tri repose sur les étapes suivantes :

- On cherche le minimum dans la liste.
- On échange ce minimum avec le premier élément de la liste.
- On recommence avec le reste de la liste, jusqu’au dernier élément.

Algorithmiquement parlant il faut :

- Commencer à chercher l’indice du minimum de la liste L; 
- Permuter le 1er élément de la liste avec l’élément minimum trouvé;
- Chercher l’indice du minimum des éléments de la sous liste L[1:]et le permuter avec le 2ème élément...
- Continuer ce principe jusqu’à ce que la liste devienne triée.

D’une manière générale, on échange l’élément à la position i avec le minimum de la sous liste L[i+1:]

Exemple:

1) On souhaite appliquer le tri par sélection à la liste L : L = [5,1,8,4,9,6,7,2,3]

- Etape N°1: Le minimum de L est 1, on le permute avec 5: L =[<font color="blue">1</font>,**5**,8,4,9,6,7,2,3]; 

- Etape N°2: Le minimum de la liste restante à trier [5,8,4,9,6,7,2,3] est 2, on le permute avec 5: L =[<font color="blue">1,2</font>,8,4,9,6,7,**5**,3]

- Etape N°3: Le minimum de la liste restante à trier  [8,4,9,6,7,5,3] est 3, on le permute avec 8: L =[<font color="blue">1,2,3</font>,4,9,6,7,5,**8**]

- Etape N°4: Le minimum de la liste restante à trier  [4,9,6,7,5,8] est 4, il est à sa bonne place: L =[<font color="blue">1,2,3,4</font>,9,6,7,5,8]

- Etape N°5: Le minimum de la liste restante à trier  [9,6,7,5,8] est 5, on le permute avec 9: L =[<font color="blue">1,2,3,4,5</font>,6,7,**9**,8]

- Etape N°6: Le minimum de la liste restante à trier  [6,7,9,8] est 6, il est à sa bonne place: L =[<font color="blue">1,2,3,4,5,</font>,7,9,8]

- Etape N°7: Le minimum de la liste restante à trier  [7,9,8] est 7, il est à sa bonne place: L =[<font color="blue">1,2,3,4,5,6,7</font>,9,8]

- Etape N°8: Le minimum de la liste restante à trier  [9,8] est 8, on le permute avec 9: L =[<font color="blue">1,2,3,4,5,6,7,8</font>,**9**]

- La liste finale triée est L =[<font color="blue">1,2,3,4,5,6,7,8,9</font>]


2) Voici une animation représentant le déroulement du tri par sélection sur la liste précédente L = [5,1,8,4,9,6,7,2,3] :



```python
def selection_sort(array):
    n = len(array)
    for i in range(0, n-1):
        minpos = i
        for j in range(i+1, n):
            if array[j] < array[minpos]:
                minpos = j
        array[i], array[minpos] = array[minpos], array[i]
        print (array)
    return array

liste_triee=selection_sort([34, 52, 1, 7, 40])
```

### Complexité du tri par sélection

Dans le pire cas ou en moyenne, la complexité ou le nombre de comparaisons du tri par sélection est en $O(n^{2})$.

## Tri fusion

En anglais : merge sort.

Ce tri a été inventé par John von Neumann en 1945. Il utilise le principe "diviser pour régner". 

L’algorithme se décrit simplement de manière récursive :
- Si la liste a 0 ou 1 élément, elle est déjà triée,
- Si la liste a plus d’un élément, on la partage en deux liste et on applique le tri fusion sur chacune des deux listes,
- On fusionne les résultats.

Pour fusionner deux listes triées, on compare les éléments de chacune des deux listes et on déplace leplus petit dans une nouvelle liste. Quand une des deux listes est vide, on déplace les éléments restants de la seconde liste.

Exemple :


 


```python
def fusion(liste1,liste2):
    liste=[]
    i,j=0,0
    while i<len(liste1)and j<len(liste2):
        if liste1[i]<=liste2[j]:
            liste.append(liste1[i])
            i+=1
        else:
            liste.append(liste2[j])
            j+=1
    while i<len(liste1):
        liste.append(liste1[i])
        i+=1
    while j<len(liste2):
        liste.append(liste2[j])
        j+=1
    return liste

def tri_fusion(liste):
    if len(liste)<2:
        return liste[:]
    else:
        milieu=len(liste)//2
        liste1=tri_fusion(liste[:milieu])
        liste2=tri_fusion(liste[milieu:])
        return fusion(liste1,liste2)

L=[3,4,6,2,5,1,8,7]
print(tri_fusion(L))
```

### Complexité du tri fusion

La complexité en temps de cet algorithme est en $O(n\log{}n)$ dans le meilleur des cas,dans le pire des cas et donc également en moyenne (dans le cas moyen).


## Tri à bulles

En anglais : bubble sort ou sinking sort.

Ce tri appelé également tri par propagation. Il consiste à comparer répétitivement les éléments consécutifs d’une liste, et à les permuter lorsqu’ils sont mal triés. Intuitivement, on fait "remonter" (comme des bulles) les éléments les plus grands jusqu’à ce qu’ils atteignent leur place.

Dans l’algorithme de tri à bulle, la liste à trier est parcourue du premier au dernier élément. L’élément courant d'indice i est comparé à l’élément suivantcelui d'indice i+1. Si l’élément en cours (d'indice i) est supérieur à l’élément suivant (d'indice i+1), il est échangé.

Exemple :

Voici une animation représentant le déroulement du tri à bulle sur la liste L = [ 6 , 5 , 3 , 1 , 8 , 7 , 2 , 4 ] :


```python
def tri_bulle(tab):
    n = len(tab)
    # Parcourir tous les éléments du tableau
    for i in range(n):
        for j in range(0, n-i-1 ):
            # échanger si l'élément trouvé est plus grand que le suivant
            if tab[j] > tab[j+1] :
                tab[j], tab[j+1] = tab[j+1], tab[j]
            print(tab)
# Programme principale pour tester le code ci-dessus
tab = [98, 22, 15, 32, 2, 74, 63, 70]
tri_bulle(tab)
```

### Complexité du tri à bulles

- Le pire cas (n itérations) est atteint lorsque le plus petit élément est à la fin du tableau. La complexité est alors $n^{2}$.
- En moyenne, la complexité est aussi $n^{2}$. En effet, le nombre d'échanges de paires d'éléments successifs est égal au nombre d'inversions de la permutation, c'est-à-dire de couples (i,j) tels que i < j et T(i) > T(j). Ce nombre est indépendant de la manière d'organiser les échanges. Lorsque l'ordre initial des éléments du tableau est aléatoire, il est en moyenne égal à $n(n-1)/4 \simeq n^{2}$.
- Le meilleur cas (une seule itération) est atteint quand le tableau est déjà trié. Dans ce cas, la complexité est linéaire $ \simeq n$.
